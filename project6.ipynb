```python
from nltk.sem.logic import Tokens
import nltk
from nltk.tokenize import word_tokenize
from nltk import pos_tag

nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')
nltk.download('punkt_tab') # Added to fix the LookupError
nltk.download('averaged_perceptron_tagger_eng') # Added to fix the new LookupError

sentence = "Natural language processing helps computer understand humans"

tokens = word_tokenize(sentence)

tags = pos_tag(tokens)

print("POS tags : ", tags)
```
